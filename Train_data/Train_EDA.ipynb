## Filtering the Dataset

```python
# Transforming the 1's and 0's into YES and NO
def Check_Yes_No(i):
    if i > 0:
        return "YES"
    else:
        return "NO"
```

```python
# Applying the funtion
data["Reached_on_Time"] = data["Reached.on.Time_Y.N"].apply(lambda i: Check_Yes_No(i))
data["Reached_on_Time"].value_counts()
```

```python
# Creating a function to convert Weight in grams to a kilgrams Class
def kgs(gms):
    Kilo_grams = round((gms*0.001),0)
    return Kilo_grams
```

```python
data["Weight_in_kgs"] = kgs(df["Weight_in_gms"])
data["Weight_in_kgs"].head()
```

Taking the columns "Reached.on.Time_Y.N" and "Customer_rating" to understand customer satisfactory levels

### Customer rating vs Reached on Time

```python
pd.crosstab(index=data["Reached_on_Time"], columns=data.Customer_rating)
```

It appears from the table above shows an equal distributed value between each rating, as the product getting to the customer on time does not affect the customer's rating in any way. 

###  Kilogram Weight Class vs Customer care calls

```python
pd.crosstab(index=data["Weight_in_kgs"], columns=data["Customer_care_calls"])
```

i. The first obvious observation is as the weight increased the customer care call reduced, especially above the 7kg region which reduced drastically

ii. The 4.0kg-6.0kg class shows an exceptional amount of calls made within that region.

### Customer care calls vs Prior purchases

```python
dap = data.groupby("Prior_purchases").agg({"Customer_care_calls":'mean'})
```

```python
drap = pd.crosstab(index=data["Prior_purchases"], columns=data["Customer_care_calls"])
```
